```python
#Heat Capacity Homework
#Code: Rosalie Chaleunsouck

#Debye's theory of solid gives heat capacity of a solid at temp T
#given: density 6.022E28m^3
# V=1000cm^3 aluminum =1000*10^-6 m^3
#thetaD= 428K
#Plot Cv as a function of temperature range 5K-500K
#In this code I am going to call the equation from scipy

#importing necessary notebooks
from scipy.integrate import quad
from math import exp
import numpy as np

#integral function in heat capacity equation
f= lambda x:x**4*exp(x)/(exp(x)-1)**2

#Given consstants by problem
V= 1000*(10**-6) #Volume of Aluminum
P= 6.022*(10**28) #Given density
Kb= 1.381*(10**-23) #Boltzmann constant
ThetaD= 428 #Debye Temperature

#Heat capacity equation definition
def Cv(T): 
    return 9*V*Kb*P*((T/ThetaD)**3)* quad(f,0,ThetaD/T)[0]

#Placing the equation points into an array and defining a range to put into a graph
Heat=[]
Series= range(5,500,1) #from 5K to 500K

#for loop to determine Cv in the temperature range
for T in Series:
    Heat.append(Cv(float(T)))
    
    
#import notebook for graph
import matplotlib.pyplot as plt

plt.plot(Series, Heat, 'r--')
plt.title("Heat Capacity of an Aluminum Solid in Response to Temperature")
plt.xlabel("Temperature (K)")
plt.ylabel("Heat Capacity (J/K)")
plt.show()
```
