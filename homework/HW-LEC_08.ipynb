```python
#Polynomial Fitting Assignment
#Code: Rosalie Chaleunsouck


#generate data with random noise (random integers?) or a function that also contains random noise?
#I am going to do a random generator for the data
#use different polynomial functions to fit the random data
#which function gives the best fit and why

#importing necessary notebooks
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit

#to make random numbered data- importing random # generator from numpy notebook
A=100
x= np.random.randint(1000, size=A)
y= np.random.randint(1000, size=A)


xaxis= np.linspace(0,1000,A) #defining my x-axis range

#DEFINING THE FUNCTIONS FOR FIT
LinFit= np.polyfit(xaxis,y,1,full=True) #first-order fit #linear
y_1_fit= np.polyval(LinFit[0], xaxis)
plt.plot

Quadfit= np.polyfit(xaxis,y,2,full=True) #second-order fit #quadratic
y_2_fit= np.polyval(Quadfit[0], xaxis)


Cubfit= np.polyfit(xaxis,y,3,full=True) #third order fit #cubic
y_3_fit= np.polyval(Cubfit[0], xaxis)

Quarfit= np.polyfit(xaxis,y,4,full=True) #fourth-order fit #quartetic? what is the term for this?
y_4_fit= np.polyval(Quarfit[0], xaxis)

#PLOT
plt.plot(x,y, 'r.', label= 'Randomized Data')
plt.plot(xaxis, y_1_fit, 'g', label= 'Linear fit')
plt.plot(xaxis, y_2_fit, 'y', label= 'Quadratic fit')
plt.plot(xaxis, y_3_fit, 'm', label= 'Cubic fit')
plt.plot(xaxis, y_4_fit, 'b', label= 'Quart fit')
plt.legend()
plt.show()

#CALCULATING THE RESIDUALS TO SEE FIT
print("Linear fit Residual", LinFit[1])
print("Quadratic fit Residual", Quadfit[1])
print("Cubic fit Residual", Cubfit[1])
print("Quart fit Residual", Quarfit[1])

#THE 4TH ORDER HAS THE LOWEST RESIDUAL FIT THUS INIDCATING THE BETTER FIT BUT SUCCESSIVE FITS MAY NEED TO BE USED TO GAUGE BETTER
```
