```python
#Decay Homework
#Code: Rosalie Chaleunsouck

#importing necessary notebooks
import numpy as np
import matplotlib.pyplot as plt

#defining my constants
tau=3.053*60
N_Atom=1000 #1000 iterations of random numbers
N_time=1000 #defining time range for decay

#defining my functions in regards to the problem
mu= np.log(2)/tau

#defining the array of number of decays in each step
N_Pb= np.zeros([N_time-1])
N_Tl= N_Atom*np.ones([N_time-1])

for i in range (N_Atom):
    z0=np.random.random()
    t=int(-np.log(1-z0)/mu) #time decay from random number
    if t<N_time:
        N_Pb[t:] += 1
        N_Tl[t:] -= 1

# creating the plot
plt.plot(N_Tl,'r-', label='Tl atoms')
plt.plot(N_Pb, 'b-', label='Pb atoms')
plt.title("Decay of Tl and Pb Over Time")
plt.xlabel("Time (s)")
plt.ylabel("# of atoms")

plt.legend()
plt.show()

```
